```python
# Plotting the histogram for CO2 distribution
plt.hist(df["co2"], bins=8)

# Adding title and labels
plt.title("CO2 Distribution")
plt.xlabel("CO2 Levels")
plt.ylabel("Frequency")

# Saving the figure
plt.savefig("co2_distribution_histogram.png")

# Displaying the plot
plt.show()
```

<h1>Plot Interpretation:</h1>
The histogram shows the distribution of CO2 levels, where most data points fall between approximately 200 and 500 CO2 units. The distribution is roughly symmetrical, with a peak (mode) around 300 to 350 CO2 units, indicating that most of the observed CO2 levels are in this range. The frequency decreases on either side of this peak, with fewer data points representing very low (under 100) or very high (over 600) CO2 levels, suggesting that extreme values are less common. The distribution appears fairly normal, but with a slight right skew, indicating a few higher CO2 levels in the dataset.

```python
# Plotting the histogram for Comb distribution
plt.hist(df["comb"], bins=18, color="black")

# Adding title and labels
plt.title("Comb Distribution")
plt.xlabel("Comb Score")
plt.ylabel("Frequency")

# Saving the figure with a more descriptive filename
plt.savefig("comb_distribution_histogram.png")

# Displaying the plot
plt.show()
```

<h1>Plot Interpretation:</h1>
The histogram shows the distribution of "Comb Score," where the majority of data points cluster between 15 and 40. The distribution is slightly right-skewed, with a peak (mode) around 25-30, indicating that most scores are concentrated in this range. The frequency decreases as the Comb Score increases beyond 30, with relatively few values above 40, suggesting that higher Comb Scores are less common in the dataset. There is a long tail on the right side, indicating the presence of a few high scores up to 60.

```python
# Plotting the histogram for feScore distribution
plt.hist(df["feScore"], bins=10, color="grey")

# Adding title and labels
plt.title("FE Distribution")
plt.xlabel("FE Score")
plt.ylabel("Frequency")

# Saving the figure with a more descriptive filename
plt.savefig("feScore_distribution_histogram.png")

# Displaying the plot
plt.show()
```

There is a slight tail on the right side, indicating the presence of a few higher scores up to 10.

```python
#data = pd.read_csv('fuel_econ.csv')
data = pd.read_csv('C:/Users/HP/Downloads/fuel_econ.csv')

# Select numeric columns (int64, float64)
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Create a correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap of Dataset')
plt.savefig("heatmap.png")
plt.show()
```

<h1>Interpreting the Data:</h1>
The heatmap visualizes correlations between the various numeric columns of the dataset. Correlations range from -1 to 1:

*   1 indicates a perfect positive correlation (as one variable increases, the other also increases).
*   -1 indicates a perfect negative correlation (as one variable increases, the other decreases).
*   0 indicates no correlation.

In this case, were focusing on the ghgScore column as the target variable. The heatmap provides the following insights:



*   co2 has a strong negative correlation with ghgScore. This means that as co2 emissions increase, the ghgScore tends to decrease, which makes sense in the context of greenhouse gas emissions.
*   feScore has a strong positive correlation with ghgScore. Higher fuel efficiency scores are associated with better greenhouse gas scores.
*   comb (possibly combined fuel economy) has a positive correlation with ghgScore, suggesting that vehicles with better combined fuel economy tend to have a better ghgScore.
*   Other variables like cylinders, displ, and UCity show some level of negative correlation with ghgScore, but not as strong as co2.



<h1>Columns to Drop if the Target is the ghgScore:</h1>
If ghgScore is the target variable, you would want to drop columns that either:

1.   Are redundant due to high correlation with other features.

<h1>Justification:</h1>
